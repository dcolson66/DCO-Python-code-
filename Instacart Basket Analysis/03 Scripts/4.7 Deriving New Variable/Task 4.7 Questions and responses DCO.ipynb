1. Create else-if price_label function 

```python
def price_label(row):

  if row['prices'] <= 5:
    return 'Low-range product'
  elif (row['prices'] > 5) and (row['prices'] <= 15):
    return 'Mid-range product'
  elif row['prices'] > 15:
    return 'High range'
  else: return 'Not enough data'
```

```python
# use price_label function on df dataset
df['price_range'] = df.apply(price_label, axis=1)
```

```python
# run value_counts function to check values in new column
df['price_range'].value_counts(dropna = False)
```

Q4. When too many users make Instacart orders at the same time, the app freezes. The senior technical officer at Instacart wants you to identify the busiest hours of the day. Rather than by hour, they want periods of time labeled Most orders, Average orders, and Fewest orders. Create a new column containing these labels called busiest_period_of_day.

```python
# Review dataframe df_orders_prod_merge
df_orders_prod_merge.head()
```

```python
# checking the frequency of order hour of day 
df_orders_prod_merge['order_hour_of_day'].value_counts(dropna = False)
```

```python
# define variable busiest_period_of_day
result3 = []

for value in df_orders_prod_merge['order_hour_of_day']:
  if value>=9 and value<=16:
    result3.append("Most orders")
  elif value<=6:
    result3.append("Fewest orders")
  else:
    result3.append("Average orders")
```

```python
result3
```

```python
len(result3)
```

```python
df_orders_prod_merge['busiest_period_of_day'] = result3
```

```python
#Print frequency of new column
df_orders_prod_merge['busiest_period_of_day'].value_counts(dropna = False)
```
